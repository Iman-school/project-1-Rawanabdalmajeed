```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

weather_data = pd.read_csv("WeatherData.csv")

weather_data
```

```python
x = weather_data["Temperature(C)"]
y = weather_data["Humidity"]

plt.scatter(x,y)
plt.xlabel("Temperature(C)")
plt.ylabel("Humidity")
```

```python
from sklearn.linear_model import LinearRegression

x = weather_data["Temperature(C)"].values.reshape(-1,1)
y = weather_data["Humidity"]

Ir_model = LinearRegression()
Ir_model.fit(x,y)
y_pred = Ir_model.predict(x)

plt.scatter(x,y)
plt.xlabel("Temperature(C)")
plt.ylabel("Humidity")

plt.plot(x,y_pred)
```

```python
theta_0 = Ir_model.intercept_
theta_1 = Ir_model.coef_
theta_0, theta_1
```

```python
y_pred = Ir_model.predict(np.array([32]).reshape(1,1))
y_pred
```
